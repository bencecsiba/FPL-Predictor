```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
import joblib

```

```python
data = pd.read_csv('../Data Cleansing/cleaned_data_4.csv', index_col=0)
```

```python
target = data['target']
in_data = data.drop(['target', 'element'], axis = 1)

train_features, test_features, train_labels, test_labels = train_test_split(in_data, target, test_size = 0.25, random_state = 42)
```

```python
rf_random = RandomForestRegressor(n_estimators= 520, min_samples_split= 2, min_samples_leaf= 4,max_features= 'sqrt',max_depth= 70, bootstrap= False)

rf = RandomForestRegressor()
random_grid = {'bootstrap': [True, False],
               'max_depth': [60, 70, 80],
               'max_features': ['log2', 'sqrt'],
               'min_samples_leaf': [2, 4, 8],
               'min_samples_split': [2, 5],
               'n_estimators': [250, 520, 1024]}

#rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = -1, scoring="neg_mean_squared_error")

rf_random.fit(train_features, train_labels)
```

```python
# Use the forest's predict method on the test data
predictions = rf_random.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'points.')
```

```python
joblib.dump(rf_random, "./random_forest_checkpoints/rf_4_" + str(round(np.mean(errors), 2)) + ".joblib")
```
